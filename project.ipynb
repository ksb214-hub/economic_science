# 📌 주제

버스 크기에 따른 대기시간 및 회전율 변화 분석을 통한 최적 운영 전략

## 1️⃣ 문제 정의

버스 크기(수용 인원)에 따라

 - 한 번에 많은 사람을 태울 수 있지만 (대형 버스)
 
 - 배차 간격이 길어질 수 있음


반대로

	• 소형 버스는 자주 오지만

    • 한 번에 적게 태움


### 👉 따라서 핵심 질문

> “어떤 버스 크기가 대기시간과 운영 효율을 동시에 최적화하는가?”

## 2️⃣ 핵심 개념

### ✔ 대기시간 (Waiting Time)

• 클수록 불편

• 최소화 목표

### ✔ 회전율 (Turnover / 처리량)

• 일정 시간 동안 수송한 인원

• 최대화 목표

### 👉 즉

> 다중 목적 최적화 문제

## 3️⃣ 변수 설정 (시험/보고서 핵심🔥)

### ✔ 의사결정변수

• x : 버스 크기 (수용 인원)

• n : 버스 운행 횟수 (또는 배차 간격)

### ✔ 매개변수
• \lambda : 학생 도착률

• \mu : 단위 시간당 처리 속도

• 운영 시간

## 4️⃣ 모델 직관

### ✔ 버스 크기 ↑
• 장점: 한 번에 많이 수송 → 회전율 ↑

• 단점: 대기시간 ↑ (버스 기다리는 시간 증가)

### ✔ 버스 크기 ↓
•	장점: 대기시간 ↓

•	단점: 회전율 ↓ (많이 못 태움)

## 5️⃣ 목적함수 (핵심🔥)

### 방법 1: 가중치 방식 (추천)
> Min Z = α(평균 대기시간) - β(회전율) 

•	α: 대기시간 중요도

•	β: 수송량 중요도

## Jupyter Notebook 코드

### 🔹 1단계: 라이브러리 import

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 🔹 2단계: 데이터 설정

```python
# 버스 크기 (수용 인원)
bus_sizes = np.array([10, 20, 30, 40, 50])

# 가정 데이터 (프로젝트용)
arrival_rate = 30   # 학생 도착률 (명/시간)

# 배차 간격 (버스 크기 커질수록 길어짐)
headway = np.array([5, 7, 9, 12, 15])  # 분

# 평균 대기시간 (단순 모델: 배차간격 / 2)
waiting_time = headway / 2

# 회전율 (시간당 수송 인원)
# = 버스 크기 / 배차간격
turnover = bus_sizes / headway
```

### 🔹 3단계: 목적함수 계산

```python
# 가중치 설정
alpha = 1   # 대기시간 중요도
beta = 2    # 회전율 중요도

# 목적함수 계산
Z = alpha * waiting_time - beta * turnover

print("목적함수 값:", Z)
```

### 🔹 4단계: 그래프 시각화

#### ✔ (1) 대기시간 vs 버스 크기

```python
plt.figure()
plt.plot(bus_sizes, waiting_time, marker='o')
plt.title("Bus Size vs Waiting Time")
plt.xlabel("Bus Size")
plt.ylabel("Waiting Time")
plt.grid()
plt.show()
```

#### ✔ (2) 회전율 vs 버스 크기

```python
plt.figure()
plt.plot(bus_sizes, turnover, marker='o')
plt.title("Bus Size vs Turnover")
plt.xlabel("Bus Size")
plt.ylabel("Turnover")
plt.grid()
plt.show()
```

#### ✔ (3) 목적함수 vs 버스 크기 ⭐ (핵심)

```python
plt.figure()
plt.plot(bus_sizes, Z, marker='o')
plt.title("Bus Size vs Objective Function (Z)")
plt.xlabel("Bus Size")
plt.ylabel("Z value")
plt.grid()
plt.show()
```

### 🔹 5단계: 최적 버스 크기 찾기

```python
optimal_index = np.argmin(Z)
optimal_bus = bus_sizes[optimal_index]

print("최적 버스 크기:", optimal_bus)
```
